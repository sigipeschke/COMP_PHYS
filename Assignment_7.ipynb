## Part 1b - Deriving Temperature

Recall that we can calculate the temperature via the followig equation:

\begin{equation}
T = \frac{\partial U}{\partial S}
\end{equation}

Where $U$ is the systems energy and $S$ is the systems entropy. If we are to determine the temperature computationally, we must take the derivative analytically which can be done as follows:

\begin{equation}
\tilde{T} \approx \frac{\tilde{U}(p_3+\Delta p_3)-\tilde{U}(p_3)}{\tilde{S}(p_3+\Delta p_3)-\tilde{S}(p_3)}
\end{equation}

As such, for us to solve this integral we must first generate two lists:

1. System Energy ($U$)
2. System Entropy ($S$)

Luckily, in our previous section we have already determined the system energy $U$ as a function of $p_3$, so we already have list (1). That means that all we are required to do is calculate the entropy for all given $p_3$'s, which can be done using the following equation:

\begin{equation}
S = -N \left[ p_1 ln(p_1) + p_2 ln(p_2) + p_3ln(p_3)\right]
\end{equation}

As such, for us to create a list of entropies we must first compute the probabilities of $p_1$, $p_2$ and $p_3$ as a function of U. This will allow us to easily calculate the entropy because at a given value of $U$ we will know the corresponding values of $p_1,p_2,p_3$ which will allow us to easily calculate an entropy value $S$ for the given $U$. As such, allow us to start by calculating the probabilities as a function of $U$:

```
# Here I have just saved an energy list (i.e values of U) that will be used in our calculation of T
N = 10000
e = -1

# Storing values
energy_list = []
p3_list = []

# Calculating values of U for many values of p_3 between 0 and 1
for i in np.arange(0,1,0.0001):
  U = (1/2)*(1*np.sqrt(4*i-3*i**2)+3*i-2)*N*e
  energy_list.append(U)
  p3_list.append(i)
```

Again, we can calculate the values of $p_1$, $p_2$ and $p_3$ as follows:

```
# Plotting p3 as a function of U
N = 10000
e = -1

# Storing all related values
uvals = []
p3_list = []
p2_list = []
p1_list = []
summed_list = []

# Determining the values of p3, p2 and p1 from the range of internal energies that we calculate above
for u in np.arange(-N,0,1):

  # Calculating probability values
  p3 = (1/6)*(-1*np.sqrt(4-3*(u/(N*e))**2) + 3*(u/(N*e)) + 4)
  p1 = p3 - u/(N*e)
  p2 = 1 - p1 - p3
  summed = p3 + p2 + p1

  # Appending these values to lists for later usage
  p3_list.append(p3)
  p2_list.append(p2)
  p1_list.append(p1)
  uvals.append(u)
  summed_list.append(summed)
```

I have added the black line labelled "Total Probability" to show that the sum of the probabilities is equivalent to 1 as expected (as $p_1 + p_2 + p_3 = 1$ for all values of $U$). Now, allow us to go ahead and use these values, along with our entropy expression, to calculate a list of entropy values for a given value of $U$. To start, I will define a function that calculate the entropy of the system:

```
def entropy(p1,p2,p3,N):

  """

  -- How it Works --

  This functions utilizes the entropy equation given above to calculate the entropy for a given value of p3. 
  This function assumes that the values of p1 and p2 also correspond to the value of p3 (i.e they are the 
  the probabilties for a given system energy U).

  -- Arguments --

  p1 = The probability of being spin up
  p2 = The probability of being spin "sideways"
  p3 = The probability of being spin down
  N = The number of units in your spin-1 pauli paramagnet

  """

  return -N*(p1*np.log(p1) + p2*np.log(p2) + p3*np.log(p3))
```

and finally, we can use this function to create our list of entropy values!

```
# Here we are simply making an entropy list that will be used later to calculate the system temperature
N = 10000
entropy_list = []

# Calculating the entropy at a given internal energy U
for p1,p2,p3 in zip(p1_list,p2_list,p3_list): # running through all 3 probability lists simultaneously
  tropy = entropy(p1,p2,p3,N)
  entropy_list.append(tropy)
```

And there we have it, our plot is exactly as we anticipated. As such, all that is left to do is to calculate the temperature for a given $U$ and $S$. This can be done with the following function:



```
def temperature_calc3r(u_val_list,s_val_list):
  

  """

  -- How it Works --

  This function is used to calculate the temperature as has been defined above.

  -- Arguments --

  u_val_list = The energy values list
  s_val_list = The list of entropy values corresponding to the energy list

  """

  # Storing all calculated T values
  T_list = []

  # Using the afformentioned expression to calculate T
  for i in range(len(u_val_list)-1):
    T = (u_val_list[i+1]-u_val_list[i])/(s_val_list[i+1]-s_val_list[i])
    T_list.append(T)

  T_list.append(T_list[-1]) # Removing the last value of the list so that it is the same size as the other functions

  return T_list
```

We can express the probability of being in a given state $i$ within the canonical ensemble as follows:

$$p_i={n_i\over N}={A\exp(-\epsilon_i/k_BT)\over
A\sum_i\exp(-\epsilon_i/k_BT)}\equiv{\exp(-\epsilon_i/k_BT)\over Z},$$

where the **partition function**, $Z$ is defined as:

$$Z\equiv\sum_i\exp\left(-{\epsilon_i\over k_BT}\right).$$

To start, allow us to construct the spin 1/2 pauli paramagnet. That is, we have spins $\uparrow$ and $\downarrow$ with the energies corresponding to the following:

$$ \epsilon_{\pm 1}=\pm\mu_BB \:\: If \:\: B>0$$

To start, we will want to construct the probability of being spin down ($p_1$) or the probability of being spin up ($p_2$) for a given temperature $\tilde{T}$ and magnetic field $\tilde{B}$. 

## Spin-1/2 Pauli Paramagnet

### Probabilities

```
def partition_calc3r( energies , T ):

  """
  
   Remember: just have a function perform one task, do not loop through in here.
   Will make things complicated if I do.

  """
  # Stroring the energy values in a list
  NRG = energies

  # Partition function is initially zero
  Z = 0

  for e in NRG: # Calculating the individal exponentials in the partition function
    Z_e = np.exp(- (e/T) )
    Z += Z_e

  return Z # Returning the final result for the partition function
```

Now, allow us to test the partition function calculator to make sure that it is returning the expected results:

```
energies = np.array([1,-1]) # The energies that we will be passing through
z_list = []

for T in np.arange(0.1,10,0.1):
  result = partition_calc3r( energies , T )
  z_list.append(result)

print(z_list)
```

And the function behaves as expected. We Start at an extremely large value for Z at low values of T, which eventually decreases steadily until it reaches an asymtote. Recall that the purpose of the partition function is to normalize the probability distribution, so it must run through all possible energies. However, our probability calculator is only required to consider one energy at a time (i.e $\epsilon_i$). Now, allow us to create a function that will determine the probabilities of a given spin up or spin down:

```
def probability_calc3r( energies, T ):

  """


  """

  # We want to start by calculating the partition function
  Z = partition_calc3r( energies , T )

  probabilities = []

  # Calculating the probability for each given energy
  for e in energies:
    probabilities.append( np.exp(-(e/T))/Z )

  return probabilities
```

### Total Internal Energy

We can calculate the total internal energy of the system by using the following equation: 

\begin{equation}
U=Nk_BT^2{\partial\ln(Z)\over\partial T}.
\end{equation}

Where all we are required to do is to take the derivative of the partition function, which yields:

$$
\frac{\partial\ln(Z)}{\partial T} = \frac{\partial}{\partial T} ln\left[ \sum_i\exp\left(-{\epsilon_i\over k_BT}\right) \right] = \frac{1}{k_b} \sum_i \frac{\frac{\epsilon_i}{T^2}exp\left(-{\epsilon_i\over k_BT} \right)}{exp(-\frac{\epsilon_i}{k_b T})}
$$

As such, we may express the total energy as follows:

$$
U=Nk_BT^2\left[ \frac{1}{k_b} \sum_i \frac{\frac{\epsilon_i}{T^2}exp\left(-{\epsilon_i\over k_BT} \right)}{exp(-\frac{\epsilon_i}{k_b T})}\right]  = N T^2 \left[ \sum_i \frac{\frac{\epsilon_i}{T^2}exp\left(-{\epsilon_i\over k_BT} \right)}{exp(-\frac{\epsilon_i}{k_b T})}\right] 
$$

As such, allow us to make a unitless function with this result:

```
def derivative_calc3r(energies,T):

  """
  
  This function calculates the derivative of ln(Z) with respect to T so that
  our energy function can be generalized for both the spin-1/2 and spin-1 pauli
  paramagnets.

  """

  numerator = 0
  denominator = 0

  for NRG in energies:
    numerator += (NRG*np.exp(-(NRG/T)))/(T**2)
    denominator += np.exp(-(NRG/T))

  return numerator/denominator
```

And finally, we can calculate the energy with our new derivative function:

```
def en3rgy_calc3r( N , energies , T ):

  """
  
  -- How it Works --

  This function simply calculates the energy of the pauli-paramagnet

  -- Arguments --

  N = Number of electrons (units)
  energies = The energies of the spins
  T = The temperature of the paramagnet
  
  """

  dZ = derivative_calc3r(energies,T)

  return N*(T**2)*dZ # Calculating the final energy expresion
```

### Entropy

Finally, we can calculate the entropy by using the following equation: 

\begin{equation}
S=Nk_B\ln(Z)+{ U\over T}.
\end{equation}

Turning this into a unitless function yields:

```
def entropy_calc3r( N , energies , T ):

  """


  """
  Z = partition_calc3r( energies , T )
  U = en3rgy_calc3r( N , energies , T )

  return N*np.log(Z) + U/T # Returning the expected entropy values
```

The spin-1 Pauli Paramagnet will be easy for us to calculate as all we are required to do is pass our functions 3 energies instead of two. These energies will be: -1 , 0 and 1. Plotting these functions, we find:

```
# All we are doing here is saving values for plotting!

energies = np.array([-1,0,1]) # The energies that we will be passing through
N_val = 10000

p_list1 = []
p_list2 = []
p_list3 = []
energy_1_list = []
entropy_1_list = []
T_list = []

# Here we are calculating the probabilities of being spin up or down at a given temperature
for T in np.arange(0.1,100,0.01):

  # -- Deriving probability lists --
  result1 = probability_calc3r( energies , T )[0] # Indicating spin down
  result2 = probability_calc3r( energies , T )[1] # Indicating spin sideways
  result3  = probability_calc3r( energies , T )[2] # Indicating spin sideways
  p_list1.append(result1)
  p_list2.append(result2)
  p_list3.append(result3)

  # Calculating Energy Probabilities 
  nrg_result = en3rgy_calc3r( N_val , energies , T )
  energy_1_list.append(nrg_result)

  # Calculating Entropy Probabilities
  ent_result = entropy_calc3r( N_val , energies , T )
  entropy_1_list.append(ent_result)

  # Appending Temperature values to plot against
  T_list.append(T)
```
